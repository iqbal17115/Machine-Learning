```python
import pandas as pd

emp=pd.read_csv('employee.csv', parse_dates=["Start Date", "Last Login Time"])
emp['Senior Management']=emp['Senior Management'].astype('bool')
emp['Gender']=emp['Gender'].astype('category')
emp.sort_values('First Name', inplace=True)
emp.head()
```

```python
len(emp)
```

```python
emp.drop_duplicates()
```

```python
len(emp.drop_duplicates())
```

```python
emp.drop_duplicates(subset=["First Name"], keep="last")
```

```python
len(emp.drop_duplicates(subset=["First Name"], keep="last"))
```

```python
emp.drop_duplicates(subset=["First Name"], keep="first")
```

```python
len(emp.drop_duplicates(subset=["First Name"], keep="first"))
```

```python
emp.drop_duplicates(subset=["First Name", "Team"], keep=False)
```

```python
len(emp.drop_duplicates(subset=["First Name", "Team"], keep=False))
```
